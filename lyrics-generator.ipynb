```
from pathlib import Path
from textgenrnn import textgenrnn

path = Path('./')
data_path = path/'drive'/'My Drive'/'generation'/'forro-data.txt'
```

# Training

```
textgen = textgenrnn(name="new_model")
textgen.reset()
textgen.train_from_file(data_path, num_epochs=10, 
                        gen_epochs=2, new_model=True, 
                        word_level=True, rnn_bidirectional=True,
                        rnn_size=64, dim_embeddings=300)
```

# Using

```
for t in textgen.generate(n=1, temperature=0.5, return_as_list=True)[0].split('#'):
  text = t.strip()
  if len(text) > 0:
    print(text.capitalize())
```
